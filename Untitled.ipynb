```python
data_obj = []
for each in data:
    if data[each].dtype == "object":
        data_obj.append(each)
```

```python
data_obj[0]
```

```python
for i in range(len(data_obj)):
    sb.catplot(y = data_obj[i], data = data, kind = "count")
```

```python
inttype = []
for each in data:
    if data[each].dtype == "int64" or data[each].dtype == "float64":
        inttype.append(each)
```
