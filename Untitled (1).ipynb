```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score

```

```python
# Dropping 'Customer ID' as it's not a useful feature for prediction
X = data.drop(['Item Purchased', 'Customer ID', 'Color','Category'], axis=1)
y = data['Item Purchased']

# Identify categorical columns for one-hot encoding, excluding 'Customer ID'
categorical_cols = X.select_dtypes(include=['object', 'category']).columns

# Preprocessing for categorical data
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Bundle preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_cols)
    ],
    remainder='passthrough'  # passthrough numerical columns as is
)

# Define Random Forest model
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Bundle preprocessing and modeling code into a pipeline
clf = Pipeline(steps=[('preprocessor', preprocessor),
                      ('model', model)
                     ])

# Splitting the dataset into training (75%) and testing (25%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Training the Random Forest model
clf.fit(X_train, y_train)

# Predicting the test set results
y_pred = clf.predict(X_test)

# Calculating accuracy
accuracy = accuracy_score(y_test, y_pred)

# Outputting the accuracy
print("Accuracy on test set:", accuracy)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

continuous_variables = ['Age', 'Purchase Amount (USD)', 'Previous Purchases']
categorical_variables = ['Gender', 'Item Purchased', 'Category', 'Location', 'Size', 'Color', 
                         'Season', 'Subscription Status', 'Shipping Type', 'Discount Applied', 
                         'Promo Code Used', 'Payment Method', 'Frequency of Purchases']

# Plot for Continuous Variables
for variable in continuous_variables:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=data, x=variable, y='Review Rating', alpha=0.6, edgecolor=None)
    plt.title(f'Review Rating vs {variable}')
    plt.xlabel(variable)
    plt.ylabel('Review Rating')
    plt.show()

# Plot for Categorical Variables
for variable in categorical_variables:
    plt.figure(figsize=(14, 8))
    sns.boxplot(data=data, x=variable, y='Review Rating')
    plt.title(f'Review Rating for each {variable}')
    plt.xticks(rotation=45)
    plt.xlabel(variable)
    plt.ylabel('Review Rating')
    plt.show()
```

```python
# Summary count of items purchased
item_counts = data['Item Purchased'].value_counts().reset_index()
item_counts.columns = ['Item Purchased', 'Purchase Count']

# Merge this count back into the main dataframe for plotting
data_with_count = pd.merge(data, item_counts, on='Item Purchased')

# Example Bar Chart for Categorical variables against Purchase Count
for variable in categorical_variables:
    if variable != 'Item Purchased':  # To avoid plotting the item purchased against itself
        plt.figure(figsize=(14, 8))
        sns.barplot(data=data_with_count, x=variable, y='Purchase Count', estimator=sum, ci=None)
        plt.title(f'Purchase Count for each {variable}')
        plt.xticks(rotation=45)
        plt.xlabel(variable)
        plt.ylabel('Purchase Count')
        plt.show()
```
